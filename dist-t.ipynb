```python
#                           Gera a distribuição t para diferentes tamanhos de amostra

import random as r
import numpy as np
import matplotlib.pyplot as plt


media = 0         # média da gaussiana geradora
dp = 1            # desvio padrão da gaussiana geradora
tamanho = 5       # tamanho da amostra
				# mude o tamanho da amostra para verificar que com n grande a distribuição t tende a ser normal.

lista = []
for i in range(int(1e6)):
  sample = np.random.normal(media, dp, size=tamanho)    # sample é um array de 5 números aleatórios (gaussianos)
  m = np.mean(sample)                                   # calcula a média da amostra
  s = np.std(sample, ddof=1)                            # calcula o desvio padrão da amostra (div por n-ddof)
  t = (m-media)/(s/np.sqrt(tamanho))                    # valor t
  lista.append(t)


hist = np.histogram(lista, bins=100, range=(-25, 25))       # tupla   ***atenção

freq, bin = hist                                            # desempacota a tupla

# Plotting
plt.figure(figsize=(10, 4))
plt.plot((bin[:-1] + bin[1:]) / 2, freq, linestyle='-', marker='o')  # plota freq versus o centro dos bins
plt.title("Distribuição t para tamanho de amostra = {}".format(tamanho))
plt.xlabel("Valor de t")
plt.ylabel("Frequência")
plt.yscale('log', base=10)
plt.grid(True)
plt.show()

freq_t = sum(freq)
print(freq_t, 'dados plotados de', int(1e6), 'faltam  ',int(1e6)-freq_t )

for i in range(100):                                        # mesmo num de bins   ***atenção
    print("%3.1f" % ((bin[i]+bin[i+1])/2), '  ',  freq[i] )   # gera dados prontos para copiar para o Prism
```
